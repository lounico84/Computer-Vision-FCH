## Ballbesitz pro Team

```python
import pandas as pd

df = pd.read_csv("project/Computer-Vision-FCH/analytics/frame_events.csv")
fps = 60.0

# 1) Gesamt-Ballbesitz pro Team (Sekunden & %)
team_frames = df[df["team_ball_control"].isin([1, 2])]
total_frames = len(team_frames)

summary = (
    team_frames["team_ball_control"]
    .value_counts()
    .rename({1: "team1", 2: "team2"})
)

team1_frames = summary.get("team1", 0)
team2_frames = summary.get("team2", 0)

team1_sec = team1_frames / fps
team2_sec = team2_frames / fps
total_sec = (team1_frames + team2_frames) / fps

team1_pct = team1_frames / (team1_frames + team2_frames) * 100 if (team1_frames + team2_frames) > 0 else 0
team2_pct = team2_frames / (team1_frames + team2_frames) * 100 if (team1_frames + team2_frames) > 0 else 0

print(f"Team 1: {team1_sec:.1f}s ({team1_pct:.1f}%)")
print(f"Team 2: {team2_sec:.1f}s ({team2_pct:.1f}%)")
```

## Ballbesitz pro Spieler

```python
# Frames, in denen jemand den Ball hat
owns = df[df["ball_owner_id"] != -1]

# Anzahl Frames mit Besitz pro Spieler
player_possession_frames = owns.groupby("ball_owner_id")["frame"].count().sort_values(ascending=False)
print(player_possession_frames.head())
```

## Ballkontakte

```python
df["owner_change"] = df["ball_owner_id"].ne(df["ball_owner_id"].shift())
touches = df[(df["owner_change"]) & (df["ball_owner_id"] != -1)]

touches_per_player = touches.groupby("ball_owner_id")["frame"].count().sort_values(ascending=False)
print(touches_per_player.head())
```

## Dauer Passketten im Team

```python
passes = touches[
    (touches["ball_owner_team"] == touches["ball_owner_team"].shift())
    & (touches["ball_owner_id"] != touches["ball_owner_id"].shift())
]

print(passes[["time_sec", "ball_owner_id", "ball_owner_team"]].head())
```
